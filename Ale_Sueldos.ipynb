### Inicio del analisis para el tratamiento de las columnas Salario_annual / Compensaciones / Moneda / Otra_Moneda

En el contexto de nuestro dataset, los participantes al estudio de salario realizado tenían que indicar sus sueldos anualizados, las compensaciones eventuales y la moneda en la cual cobraban. El valor de la columa Moneda es una lista de 11 monedas. En el caso en el que el participante cobre en una moneda ausente de la lista de 11 monedas, tiene la posibilidad de eligir manualmente la moneda del cobro

```python
#Conocer los valores unicos de col. Moneda
df["Moneda"].unique()
```

```python
#Conocer la redudancia de los valores unicos de col. Moneda
df["Moneda"].value_counts()
```

```python
#Conocer los valores unicos de col. Otra_Moneda
df["Otra_Moneda"].unique()
```

```python
#Conocer la redudancia de los valores unicos de col. Otra_Moneda
df["Otra_Moneda"].value_counts()
```

Según lo analizado hasta áca, muchos "otras monedas" se repiten pero algunos datos ingresados no son relevantes para el estudio (aparecen sólo una vez, los datos ingresados no tienen contexto o valor). Se decide entonces tomar en cuenta las "otras monedas" que aparecen 5 veces o más. El resto puede ser eliminados porque son datos no concluyentes

```python
#Filtrar donde la moneda es Other
otras_monedas = df[df["Moneda"] == "Other"]

#Contar los valores de Other en Moneda
conteo_otras_monedas = otras_monedas["Otra_Moneda"].value_counts()

#Filtrar las monedas comunes con mas de 5 apariciones
Others_mas_comunes = conteo_otras_monedas[conteo_otras_monedas > 4]

#Ver el resultado
Others_mas_comunes
```

Se decide normalizar todos los sueldos en dolares para poder analizarlos

```python
#Normalizar sueldos en dolares 

paso_a_usd = {

    "USD": 1,
    "CAD": 0.74,
    "GBP": 1.25,
    "EUR": 1.08,
    "AUD/NZD": 0.66,
    #"Other": 1,
    "CHF": 1.12,
    "SEK": 0.098,
    "JPY": 0.0066,
    "ZAR": 0.054,
    "HKD": 0.13,
    "INR": 0.012,
    "SGD": 0.74,
    "MYR": 0.21,
    "NOK": 0.091,
    "USD": 1,
    "DKK": 0.14,
    "BRL": 0.20,
    "PLN": 0.25
    
}

#Funciones para convertir

def convertir_salario_en_usd(fila):
    cambio = paso_a_usd.get(fila["Moneda"], 1)
    return fila["Salario_Anual"] * cambio

def convertir_compensation_en_usd(fila):
    cambio = paso_a_usd.get(fila["Moneda"], 1)
    return fila["Compensaciones"] * cambio

def convertir_others_en_usd(fila):
    cambio = paso_a_usd.get(fila["Otra_Moneda"], 1)
    return fila["Salario_Anual"] * cambio

def convertir_others_compensacion_en_usd(fila):
    cambio = paso_a_usd.get(fila["Otra_Moneda"], 1)
    return fila ["Compensaciones"] * cambio

# Aplicamos a filas donde la moneda es distinta de "Other"
df.loc[df["Moneda"] != "Other", "Salario_en_USD"] = df[df["Moneda"] != "Other"].apply(convertir_salario_en_usd, axis=1)
df.loc[df["Moneda"] != "Other", "Compensaciones_en_USD"] = df[df["Moneda"] != "Other"].apply(convertir_compensation_en_usd, axis=1)

# Aplicamos a filas donde la moneda es "Other" usando la columna Otra_Moneda
df.loc[df["Moneda"] == "Other", "Salario_en_USD"] = df[df["Moneda"] == "Other"].apply(convertir_others_en_usd, axis=1)
df.loc[df["Moneda"] == "Other", "Compensaciones_en_USD"] = df[df["Moneda"] == "Other"].apply(convertir_others_compensacion_en_usd, axis=1)

df["Remuneración_total_en_USD"] = df["Salario_en_USD"] + df["Compensaciones_en_USD"]


df
```

```python
# Filas "Other" con monedas no frecuentes para su posterior eliminación
otros_no_utiles = df[(df["Moneda"] == "Other") & (~df["Otra_Moneda"].isin(Others_mas_comunes.index))]

# Eliminarlas del DataFrame original
df = df.drop(otros_no_utiles.index)
df
```

### Analisis grafícas de nuestra primera limpeza

```python
#descarga de una librería de gráfica y generación de un gráfico de dispersón de sueldos para su posterior analisis

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(df["Experiencia_Total"], df["Remuneración_total_en_USD"], alpha=0.5)
plt.title("Dispersión de salarios según experiencia total")
plt.xlabel("Años de experiencia")
plt.ylabel("Salario total en USD")
plt.grid(True)
plt.show()
```

Notamos importantes outliers, que dificultaran el uso de los datos
A continuación, hacemos una analisis más detallado

```python
#bFiltro de los 10 sueldos más altos
top_10_sueldos = df.sort_values(by="Remuneración_total_en_USD", ascending=False).head(10)
top_10_sueldos
```

```python
#Filtro de los 10 sueldos más bajos
flop_10_sueldos = df.sort_values(by="Remuneración_total_en_USD", ascending=True).head(10)
flop_10_sueldos
```

En este momento, notamos que algunos sueldos están demasiado alto y sin contexto y que otros sueldos están a 0 o muy bajos. Interpretamos que son un error de carga de datos.
A continuación, eliminamos los sueldos a 0 y eliminamos los sueldos por debajo del percentil 5 y arriba del percentil 95 para dar más consistencia a nuestro dataset

```python
#Filtrar las personas que tienen cómo sueldo total 0 para su posterior eliminación
Rem_total_cero = df[df["Remuneración_total_en_USD"] == 0]
Rem_total_cero
```

```python
#Eliminar los sueldos a 0
df = df[df["Remuneración_total_en_USD"] > 0]
```

```python
#Analisis de los percentiles 5 y 95 para posterior eliminación de Outliers

p5 = df["Remuneración_total_en_USD"].quantile(0.05)
p95 = df["Remuneración_total_en_USD"].quantile(0.95)

#Modificación del dataset para tomar en cuenta los datos entre percentil 5 y percentil 95

df_sin_outliers = df[(df["Remuneración_total_en_USD"] >= p5) & (df["Remuneración_total_en_USD"] <= p95)]
```

```python
#Generación de un gráfico de dispersión

plt.figure(figsize=(10, 6))
plt.scatter(df_sin_outliers["Experiencia_Total"], df_sin_outliers["Remuneración_total_en_USD"], alpha=0.5)
plt.title("Dispersión de salarios según experiencia total")
plt.xlabel("Años de experiencia")
plt.ylabel("Salario total en USD")
plt.grid(True)
plt.show()
```

Vemos una mejor dispersión. A continuación, vamos a mejorar el orden de nuestro eje X y proponer un gráfico de bigotes para cada rango etario

```python
#Código para mejorar el orden de nuestro eje X

orden_experiencia = [
    '1 year or less',
    '2 - 4 years',
    '5-7 years',
    '8 - 10 years',
    '11 - 20 years',
    '21 - 30 years',
    '31 - 40 years',
    '41 years or more'
]

df_sin_outliers["Experiencia_Ordenada"] = pd.Categorical(
    df_sin_outliers["Experiencia_Total"],
    categories=orden_experiencia,
    ordered=True
)

```

```python
#Generación de un gráfico boxplot

import seaborn as sns
plt.figure(figsize=(12, 6))
sns.boxplot(x="Experiencia_Ordenada", y="Remuneración_total_en_USD", data=df_sin_outliers)
plt.xticks(rotation=45)
plt.title("Distribución de salarios según años de experiencia")
plt.xlabel("Años de experiencia")
plt.ylabel("Salario total en USD")
plt.grid(True)
plt.show()
```
